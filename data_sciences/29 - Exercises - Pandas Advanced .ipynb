```python
# Inner merge the 2 data sets based on key

data1 = pd.DataFrame({'key': ['K0', 'K1', 'K1', 'K2'],
                     'P': ['P0', 'P1', 'P2', 'P3'],
                     'Q': ['Q0', 'Q1', 'Q2', 'Q3']}) 
data2 = pd.DataFrame({'key': ['K0', 'K0', 'K1', 'K3'],
                      'R': ['R0', 'R1', 'R2', 'R3'],
                      'S': ['S0', 'S1', 'S2', 'S3']})




data1.merge(data2, on='key', how='inner')



# Type your code above this line
```

```python
# Outer merge the 2 data sets based on key

data1 = pd.DataFrame({'key': ['K0', 'K1', 'K1', 'K2'],
                     'P': ['P0', 'P1', 'P2', 'P3'],
                     'Q': ['Q0', 'Q1', 'Q2', 'Q3']}) 
data2 = pd.DataFrame({'key': ['K0', 'K0', 'K1', 'K3'],
                      'R': ['R0', 'R1', 'R2', 'R3'],
                      'S': ['S0', 'S1', 'S2', 'S3']})


data1.merge(data2, on='key', how='outer')


# Type your code above this line
```

```python
# Left merge the 2 data sets based on key

data1 = pd.DataFrame({'key': ['K0', 'K1', 'K1', 'K2'],
                     'P': ['P0', 'P1', 'P2', 'P3'],
                     'Q': ['Q0', 'Q1', 'Q2', 'Q3']}) 
data2 = pd.DataFrame({'key': ['K0', 'K0', 'K1', 'K3'],
                      'R': ['R0', 'R1', 'R2', 'R3'],
                      'S': ['S0', 'S1', 'S2', 'S3']})


data1.merge(data2, on='key', how='left')


# Type your code above this line
```
