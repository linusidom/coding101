```python
# Print numbers -1 to -10 and -10 to -1

# While Loop

print('While Loop Decrement')
count = -1 # Start Case
while count > -11: # Stop
    print(count)
    count -= 1 # Step Case

print('\nWhile Loop Increment')
count = -10 # Start Case
while count < 0: # Stop
    print(count)
    count += 1 # Step Case
    
print('\nFor Loop Decrement')

for count in range(-1, -11, -1):
    print(count)

print('\nFor Loop Increment')

for count in range(-10, 0, 1):
    print(count)
    
  
print('\nList Comprehension Decrement')    
[print(count) for count in range(-1,-11,-1)]

print('\nList Comprehension Increment')    
[count for count in range(-10,0,1)]


# Type your code above this line
```

```python
# Print numbers 1 to 10 and 10 to 1

# While loop increment
print('While Loop Increment')
count = 1
while count < 11:
    print(count)
    count += 1

print('\nWhile Loop Decrement')
count = 10
while count > 0:
    print(count)
    count -= 1

print('\nFor Loop Increment')
for count in range(1,11):
    print(count)

print('\nFor Loop Decrement')
for count in range(10,0, -1):
    print(count)

print('\nList Comp Increment')
[print(count) for count in range(1,11)]

print('\nList Comp Decrement')
[print(count) for count in range(10,0,-1)]
# Type your code above this line
```

```python
# Mutliply only the odd numbers between 0 and 20 by 2
# Return the results in a list called list_odds

# While

list_odds = []
count = 0
while count < 21:
    if count % 2 == 1: # This identifies odd numbers
#         print(count * 2)
        list_odds.append(count * 2)
    count += 1
print(list_odds)

list_for_odds = []

for count in range(21):
    if count % 2 == 1:
        list_for_odds.append(count * 2)
print(list_for_odds)


# List Comprehension
list_comp_odds = [count * 2 for count in range(21) if count % 2 == 1]
print(list_comp_odds)

# Type your code above this line
```
