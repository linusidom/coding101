### Creating Charts

2 Methods, Functional and Object Oriented

###### Functional Method

If we only need a single chart we can use this method, faster and easier to use, but limited because we overwrite our previous data

`
x = np.linspace(0,11,5)
y = x ** 2
plt.plot(x, y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
`


###### Object Oriented Method

Instantiate a `fig` object first from `plt.figure()` then we modify the element attribute by attribute

This method allows us to have more charts and call upon them are we want to

`
fig, ax = plt.subplots(1,2)
ax[0].plot([1, 2, 3, 4], [1, 4, 2, 3]); 
ax[1].plot([2, -2, -3, -4], [-1, 4, -2, 3]); 
ax[0].set_xlabel('X Label One')
ax[0].set_ylabel('Y Label One')
ax[0].set_title('Title One')
ax[1].set_xlabel('X Label Two')
ax[1].set_ylabel('Y Label Two')
ax[1].set_title('Title Two')
fig.tight_layout()
`

Note: `add_axes([Left, Bottom, Width, Height])` -> all values between 0 and 1

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Our first chart
# All components of a chart / plot must be in the same cell

x = np.linspace(0,11,5)
y = x ** 2

print(x, y)

# Function Oriented Plotting

plt.plot(x, y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')

```

```python
# Object Oriented Plotting

fig, ax = plt.subplots(nrows=1,ncols=2)

print(ax)
x=[1, 2, 3, 4]
y=[1, 4, 2, 3]

ax[0].plot(x, y); 

ax[1].plot([2, -2, -3, -4], [-1, 4, -2, 3]); 

ax[0].set_xlabel('X Label One')
ax[0].set_ylabel('Y Label One')
ax[0].set_title('Title One')

ax[1].set_xlabel('X Label Two')
ax[1].set_ylabel('Y Label Two')
ax[1].set_title('Title Two')

fig.tight_layout()
```

```python
# Working with DataFrames

df = pd.read_csv('Titanic.csv')
df.head()
```

```python
n_bins = 40
fig, axs = plt.subplots(1, 2, tight_layout=True)

# We can set the number of bins with the *bins* keyword argument.
axs[0].hist(df['Age'], bins=n_bins)

axs[0].set_xlabel('Age')
axs[0].set_ylabel('Count')
axs[0].set_title('Total Count by Age')


axs[1].hist(df['Fare'], bins=50)

axs[1].set_xlabel('Fare')
axs[1].set_ylabel('Count')
axs[1].set_title('Total Count by Fare')


```

```python
# Scatter Plot


plt.scatter(x='Age', y='Pclass', data=df)
```

```python
# Grouped Bar Chart

labels = df['Pclass'].unique()[::-1]
print(labels)

avg_age = [df[df['Pclass'] == label]['Age'].mean() for label in labels]
print(avg_age)

total_count = [df[df['Pclass'] == label]['Age'].count() for label in labels]
print(total_count)

x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, avg_age, width, label='Avg Age')
rects2 = ax.bar(x + width/2, total_count, width, label='Total Count')

# # Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Count / Age')
ax.set_title('Count and Age by Class')
ax.set_xticks(x, labels)
ax.legend()

ax.bar_label(rects1, padding=3)
ax.bar_label(rects2, padding=3)

fig.tight_layout()

# plt.show()
```
